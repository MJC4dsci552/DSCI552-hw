## Step 0 Import Required Libraries ##

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
```

## Step 1 Read and Load the Dataset ##

```python
# Load the dataset from the provided files
data = pd.read_csv('../data/column_2C.dat', sep=' ', header=None, names=['pelvic_incidence', 'pelvic_tilt', 'lumbar_lordosis_angle', 'sacral_slope', 'pelvic_radius', 'grade_spondylolisthesis', 'class'])

# Map class labels to binary values (NO=0, AB=1)
data['class'] = data['class'].map({'NO': 0, 'AB': 1})

# Create separate datasets for Class 0 and Class 1 for easliy drawing plots.
class_0_data = data[data['class'] == 0]
class_1_data = data[data['class'] == 1]
# print(data)
```

# (b)Pre-Processing and Exploratory data analysis:#

## b(i): Scatterplots of Independent Variables ##

```python
#i. Scatterplots of independent variables
# make class0 to be blue and class 1 to be red
colors = ['blue' if c == 0 else 'red' for c in data['class']]
pd.plotting.scatter_matrix(data.iloc[:, :-1], c=colors, alpha=0.6, figsize=(12, 12), marker='o', hist_kwds={'bins': 20})
plt.suptitle('Scatterplots of Independent Variables (0: NO, 1: AB)', fontsize=16)
plt.show()
```

## b(ii) Make boxplots for each of the independent variables##

```python
plt.figure(figsize=(12, 6))
for i, col in enumerate(data.columns[:-1]):
    plt.subplot(2, 3, i + 1)
    boxprops = dict(color='blue', facecolor='lightblue')
    data[data['class'] == 0].boxplot(positions = [0], column=col, patch_artist=True, boxprops=boxprops)
    boxprops = dict(color='red', facecolor='lightcoral')
    data[data['class'] == 1].boxplot(positions = [1],column=col, patch_artist=True, boxprops=boxprops)
    plt.title(col)
    plt.xticks([0, 1], ['Class 0', 'Class 1'])
plt.suptitle('Boxplots of Independent Variables (0: NO, 1: AB)', fontsize=16)
plt.tight_layout()
plt.show()
```

## b(iii)Select training set and the rest of the data as the test set. ##

```python
# iii. Select the first 70 rows of Class 0 and the first 140 rows of Class 1 as the training set
# and the rest of the data as the test set
class_0_train = class_0_data.head(70)
class_1_train = class_1_data.head(140)
training_data = pd.concat([class_0_train, class_1_train], axis=0)
test_data = data.drop(training_data.index)
```

# (c) Classification using KNN on Vertebral Column Data Set #

## c(i)(ii) Write code for k-nearest neighbors with Euclidean metric ##

```python
# Separate features (X) and labels (y)
y_train = training_data['class']
X_train = training_data.drop('class', axis=1)
y_test = test_data['class']
X_test = test_data.drop('class', axis=1)


# Define values of k to test
k_values = list(range(208, 0, -3))

# Initialize lists to store train and test errors
train_errors = []
test_errors = []

for k in k_values:
    # Create a KNN classifier with the current value of k
    knn = KNeighborsClassifier(n_neighbors=k)
    
    # Train the model
    knn.fit(X_train, y_train)
    
    # Predict on the training data
    y_train_pred = knn.predict(X_train)
    
    # Predict on the testing data
    y_test_pred = knn.predict(X_test)
    
    # Calculate training and testing errors
    train_error = 1 - knn.score(X_train, y_train)
    test_error = 1 - knn.score(X_test, y_test)
    
    # Append errors to the lists
    train_errors.append(train_error)
    test_errors.append(test_error)

# Find the index of the minimum test error
best_k_index = np.argmin(test_errors)
best_k = k_values[best_k_index]

# Plot train and test errors in terms of k
plt.figure(figsize=(10, 6))
plt.plot(k_values, train_errors, label='Train Error', marker='o')
plt.plot(k_values, test_errors, label='Test Error', marker='o')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Error Rate')
plt.title('Train and Test Errors vs. Number of Neighbors (k)')
plt.legend()
plt.grid(True)
plt.show()

# Use the best_k to create the final KNN classifier
knn_best = KNeighborsClassifier(n_neighbors=best_k)
knn_best.fit(X_train, y_train)
y_test_pred_best = knn_best.predict(X_test)

# Calculate confusion matrix, precision, recall, and F1-score
conf_matrix = confusion_matrix(y_test, y_test_pred_best)
precision = precision_score(y_test, y_test_pred_best)
recall = recall_score(y_test, y_test_pred_best)
f1 = f1_score(y_test, y_test_pred_best)
print(f'K*(Best k):{best_k}')
print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1-Score: {f1}')
```

## c(iii) Learning Curve. ##

```python
N_values = list(range(10, 210 , 10))
best_k_values = list(range(1, 200, 5))
# colors = plt.cm.viridis(np.linspace(0, 1, 210))

test_errors_curve = []

for N in N_values:
 
    # Create training set with first N/3 rows of Class 0 and first N - N/3 rows of Class 1
    train_data_class_0 = class_0_data.head(N // 3)
    train_data_class_1 = class_1_data.head(N - (N // 3))
    train_data_curve = pd.concat([train_data_class_0, train_data_class_1])
    
    X_train_curve = train_data_curve.drop('class', axis=1)
    y_train_curve = train_data_curve['class']
    
    test_error_curve = []
    
    for k in best_k_values :
        if N < k: # if N < k, we can not build a KNeighborsClassfier
            test_error_curve.append(None)
            continue 
        # Get test error    
        knn_curve = KNeighborsClassifier(n_neighbors=k)
        knn_curve.fit(X_train_curve, y_train_curve)
        y_test_pred_curve = knn_curve.predict(X_test)
        test_error_curve.append(1 - knn_curve.score(X_test, y_test))
    
    test_errors_curve.append(test_error_curve)

# Plot Learning Curve
plt.figure(figsize=(10, 6))
for i, N in enumerate(N_values):
    plt.plot(best_k_values, test_errors_curve[i], label=f'N={N}')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Test Error Rate')
plt.title('Learning Curve')
plt.legend()
plt.grid(True)
plt.show()
```

# (d) Replace the Euclidean metric with the following metrics5 and test them # 

## d(i). Minkowski Distance: A. which becomes Manhattan Distance with p = 1. ##

```python
# Define values of k to test
k_values_d = list(range(1, 196, 5))

# Initialize lists to store test errors

test_errors_d1 = []

for k in k_values_d:
    # Create a KNN classifier with the current value of k and p=1(Manhattan Distance)
    knn_d1 = KNeighborsClassifier(n_neighbors=k, p=1,metric='manhattan')
    
    # Train the model
    knn_d1.fit(X_train, y_train)
    
    
    # Predict on the testing data
    y_test_pred_d1 = knn_d1.predict(X_test)
    
    # Calculate testing errors
    test_error_d1 = 1 - knn_d1.score(X_test, y_test)
    
    # Append errors to the lists
    test_errors_d1.append(test_error_d1)

# Find the index of the minimum test error
best_k_index_d1 = np.argmin(test_errors_d1)
best_k_d1 = k_values_d[best_k_index_d1]

# Plot train and test errors in terms of k
plt.figure(figsize=(10, 6))
plt.plot(k_values_d, test_errors_d1, label='Test Error', marker='o')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Error Rate')
plt.title('Test Errors vs. Number of Neighbors (k) for Manhattan Distance')
plt.legend()
plt.grid(True)
plt.show()

print(f'K*(Best k) for Manhattan Distance:{best_k_d1}')

knn_best_d1 = KNeighborsClassifier(n_neighbors=best_k_d1, p=1)
knn_best_d1.fit(X_train, y_train)
test_error_d1 = 1 - knn_best_d1.score(X_test, y_test)
print(f'Best test error for Manhattan Distance:{test_error_d1}')


```

## d(i)B: with log10(p) ##

```python
# log10 p range in 0.1 to 1.0
log10_p_values = np.arange(0.1, 1.0, 0.1)

test_errors_d2 = []

for p1 in log10_p_values:
    # Create a KNN classifier with the current value of best_k = 6 and log10 p
    knn_d2 = KNeighborsClassifier(n_neighbors=best_k_d1, p=10**p1)
    
    # Train the model
    knn_d2.fit(X_train, y_train)
    
    
    # Predict on the testing data
    y_test_pred_d2 = knn_d2.predict(X_test)
    
    # Calculate testing errors
    test_error_d2 = 1 - knn_d2.score(X_test, y_test)
    
    # Append errors to the lists
    test_errors_d2.append(test_error_d2)

# Find the index of the minimum test error
best_p_index_d2 = np.argmin(test_errors_d2)
best_p_d2 = log10_p_values[best_p_index_d2]

# Plot train and test errors in terms of k
plt.figure(figsize=(10, 6))
plt.plot(log10_p_values, test_errors_d2, label='Test Error', marker='o')
plt.xlabel('log10 p')
plt.ylabel('Error Rate')
plt.title('Test Errors vs. log10 p')
plt.legend()
plt.grid(True)
plt.show()

print(f'Best log10 p for Minkowski Distance:{best_p_d2}')

knn_best_d2 = KNeighborsClassifier(n_neighbors=best_k_d1, p=10**best_p_d2)
knn_best_d2.fit(X_train, y_train)
test_error_d2 = 1 - knn_best_d2.score(X_test, y_test)
print(f'Best test error for Minkowski Distance:{test_error_d2}')
```

## d(i)C:  Chebyshev Distance ##

```python
# Initialize lists to store test errors

test_errors_chebyshev = []

for k in k_values_d:
    # Create a KNN classifier with the current value of k and p=inf(chebyshev Distance)
    knn_chebyshev = KNeighborsClassifier(n_neighbors=k, metric='chebyshev', p = float('inf'))
    
    # Train the model
    knn_chebyshev.fit(X_train, y_train)
    
    
    # Predict on the testing data
    y_test_pred_chebyshev = knn_chebyshev.predict(X_test)
    
    # Calculate testing errors
    test_error_chebyshev = 1 - knn_chebyshev.score(X_test, y_test)
    
    # Append errors to the lists
    test_errors_chebyshev.append(test_error_chebyshev)

# Find the index of the minimum test error
best_k_index_chebyshev = np.argmin(test_errors_chebyshev)
best_k_chebyshev = k_values_d[best_k_index_chebyshev]

# Plot train and test errors in terms of k
plt.figure(figsize=(10, 6))
plt.plot(k_values_d, test_errors_chebyshev, label='Test Error', marker='o')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Error Rate')
plt.title('Test Errors vs. Number of Neighbors (k) for chebyshev')
plt.legend()
plt.grid(True)
plt.show()

print(f'K*(Best k) for chebyshev Distance):{best_k_chebyshev}')

knn_best_chebyshev = KNeighborsClassifier(n_neighbors=best_k_chebyshev, metric='chebyshev',p = float('inf'))
knn_best_chebyshev.fit(X_train, y_train)
test_error_chebyshev1 = 1 - knn_best_chebyshev.score(X_test, y_test)
print(f'Best test error for chebyshev Distance:{test_error_chebyshev1}')
```

## d(ii) Mahalanobis Distance. ##

```python
# Initialize lists to store test errors

test_errors_mahalanobis = []

# Inverse of the covariance matrix for Mahalanobis distance
VI = np.linalg.inv(np.cov(X_train, rowvar=False))

for k in k_values_d:
    # Create a KNN classifier with the current value of k and mtric = mahalanobis

    
    knn_mahalanobis = KNeighborsClassifier(n_neighbors=k, metric='mahalanobis',metric_params={'VI': VI})   
    
    # Train the model
    knn_mahalanobis.fit(X_train, y_train)
    
    
    # Predict on the testing data
    y_test_pred_mahalanobis = knn_mahalanobis.predict(X_test)
    
    # Calculate testing errors
    test_error_mahalanobis = 1 - knn_mahalanobis.score(X_test, y_test)
    
    # Append errors to the lists
    test_errors_mahalanobis.append(test_error_mahalanobis)

# Find the index of the minimum test error
best_k_index_mahalanobis = np.argmin(test_errors_mahalanobis)
best_k_mahalanobis = k_values_d[best_k_index_mahalanobis]

# Plot train and test errors in terms of k
plt.figure(figsize=(10, 6))
plt.plot(k_values_d, test_errors_mahalanobis, label='Test Error', marker='o')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Error Rate')
plt.title('Test Errors vs. Number of Neighbors (k) for mahalanobis distance')
plt.legend()
plt.grid(True)
plt.show()

print(f'K*(Best k) for chebyshev Distance):{best_k_mahalanobis}')

knn_best_mahalanobis = KNeighborsClassifier(n_neighbors=best_k_mahalanobis, metric='mahalanobis',metric_params={'V': np.cov(X_train, rowvar=False)})
knn_best_mahalanobis.fit(X_train, y_train)
test_error_mahalanobis1 = 1 - knn_best_mahalanobis.score(X_test, y_test)
print(f'Best test error for chebyshev Distance:{test_error_mahalanobis1}')
```
